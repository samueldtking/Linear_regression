```python
# Load libraries 
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

import statsmodels.api as sm
import statsmodels.formula.api as sm
import statsmodels.api as sma
from statsmodels.tools.eval_measures import mse
from statsmodels.tools.tools import add_constant

from sklearn import linear_model,cross_validation, feature_selection,preprocessing
from sklearn.metrics import mean_squared_error
from sklearn.cross_validation import KFold
```

```python
# LINEAR REGRESSION WITH STATSMODELS 
# For this example I explicitly chose a dataset with poor correlation to demonstrate how to spot a linear regression that does not work well




# STEP 1: LOAD DATA (AND DESCRIBE)
data = pd.read_csv('soccer_fantasy.csv')
pd.set_option('precision',3)
data.describe()
```

```python
# STEP 2: CORRELATION OF VARIABLES (INTERPRET)
# It is important to start by taking time to examine and interpret correlation between each variable  

data.corr()
```

```python
# Correlation plots are a very useful way of seeing the correlation between all variables (as well as their distribution)

from pandas.tools.plotting import scatter_matrix
scatter_matrix(data,figsize=(8, 8), diagonal='kde');
```

```python
# STEP 3: PLOT 2 WAY AND 3 WAY SCATTER PLOTS BETWEEN VAIRABLES TO EXPLORE INTERESTING RELATIONSHIPS FURTHER

# For this example I am using 'Fantasy points' as the dependent variable (y)
# It is useful at an early stage to compare all variables agains this dependent variable to explore relationships 
data_x = data.columns[:-1]

for i in data_x:
    fig, ax = plt.subplots(1, 1)  
    ax.scatter(data[i], data.avg_fantasy_points)
    ax.set_xlabel(i)
    ax.set_ylabel('Average fantasy points')
    plt.show()
```

```python
# 3d plots
# it may also be useful to explore 3 way correlations (though these can often be difficult to see visually)
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = Axes3D(fig, azim=-115, elev=15)
ax.scatter(data['weight'], data['shots_on_target_rate'], 
           data['avg_fantasy_points'])
ax.set_xlabel('Weight')
ax.set_ylabel('Shots on target')
ax.set_zlabel('Avg. Fantasy Points')
plt.show()
```

```python
# STEP 4: MAKE COPY AND SPLIT INTO TRAINING/TEST/VALIDATION SET 
# tip = copy the data so you don't corrupt the original dataset 
# note need to remember to add crossvalidation here (SciKit comes with functions to automate cross validation)


X =  data.copy() 
X.columns = ['m_dash', 'weight', 'shots_on_target_rate', 'penalty_rate','avg_fantasy_points']

train,test = cross_validation.train_test_split(X,train_size=0.80)
```

```python
# STEP 5: FIT TRAINING SET AND CREATE MODEL (LOOK AT SUMMARY)
# note = in statsmodeel 'ols' is in lowercase, this is always the case for formula api (upper case for non formulla api)
# capital variant has train/copy the other way around 
# also need to remember to add the +1  to add interpcet = not by deault (in sklearn = its default)

model = sm.ols(formula= 'avg_fantasy_points ~ 1 + m_dash + weight + shots_on_target_rate + penalty_rate', 
               data=train).fit()

model.summary()

# If it isn't working check the exact names of columns and rename X columns if needed 
```

```python
# STEP 6: INTERPRETING RESULTS 
# DOC: http://www.statsmodels.org/dev/generated/statsmodels.regression.linear_model.RegressionResults.html

#TABLE 1:
# Interpretting the summary
    # R-squared: error value - close to 1 is good, far from 1 is bad
                # though weakness of normal r^2 is that it will only improve by adding more variables 
    # Adj-R-Squared: error taking into account # of variables - again close to 1 good 
                # Adjusted R^2 - normalises r^2  for the number of variables you have - 
    # F-statistic: overall significance of the model - under - 0.05 mean model IS signficinat otherwise model is no better...
                # ...than if all beta add to zero (e.g. model is no better than just using mean)
                # F statistic in the overall statistics of the overall model
                # Calculated as the mean squared error of the model divided by the mean squared error of the residuals.
                # IF F VALUE OK AND ALL P VALUES NOT = THEN VARIABLES ARE LIKELY CORRELATED WITH EACH OTHER 
    # Prob (F-statistic): see above 
                # p-value of the F-statistic
    # Log-Likelihood: higher value is better 
                # A likelihood-ratio test is a statistical test relying on a test statistic computed by taking the ratio of the maximum value of the likelihood function under the constraint of the null hypothesis to the maximum with that constraint relaxed.
    # AIC: Akaike Information Criterion - penalises additional variables - used to compare models - lower AIC is preferred 
                # The Akaike information criterion (AIC) is a measure of the relative quality of statistical models for a given set of data.
    # BIC: Bayesian Infomation Criterion - lower BIC is preferred 
                #  Bayesian information criterion (BIC) or Schwarz criterion (also SBC, SBIC) is a criterion for model selection among a finite set of models; the model with the lowest BIC is preferred
    
    
#TABLE 2 
    # Interpretting the coefficients
        # Coefficients of variables: beta coefficent of the variavle (increase for each unit of the var) (y = b + b1x1 + bnxn ...)
        # Std error of variables: standard error of the variables around the central value 
        # Pvalue>value: less than 0.05 significant otherwise more than 5% chance the correlation is supurious  
                # NOTE IF PVALUES ARE HIGH BUT MODEL IS SIGNIFICANT THAT LIKELY MULTICOLINEARITY BETWEEN VARIABLES    
    
# TABLE 3
    # Interpretting distribution statistics 
        # Omnibus: test if explained variance is bigger than unexplained variance 
        # Prob(Omnibus): p-value for omnibus stat
        # Skew: measure of asymmetry should be 0 if totally symetrical # https://www.graphpad.com/guides/prism/6/statistics/index.htm?stat_skewness_and_kurtosis.htm 
        # Kurtosis: measure of sharpness/flatness of data, should be close to 3 
        
        # Durbin-Watson: Durbin–Watson statistic is a test statistic used to detect the presence of autocorrelation (a relationship between values separated from each other by a given time lag) in the residuals (prediction errors) from a regression analysis. d = 2 indicates no autocorrelation. The value of d always lies between 0 and 4. 
        # Jarque-Bera: Jarque–Bera test is a goodness-of-fit test of whether sample data have the skewness and kurtosis matching a normal distribution. 
        # Prob(JB): Pvalue for JB stat

        # Conditional number: ABOVE 30 SUGGESTS MULTICOLINEARITY 


```

```python
# STEP 7: FIT MODEL WITH CROSS-VALIDATION 

train,test = cross_validation.train_test_split(X,train_size=0.80)

model = sm.ols(formula='avg_fantasy_points ~ 1+ m_dash + weight + shots_on_target_rate + penalty_rate', 
               data=train).fit()

modelforout = model 

model.summary()


```

```python
# STEP 8: CHECK LINEAR REGRESSION ASSUMPTIONS

# Linear regression Assumptions 

#1. Linear relationship
        # check outliers (linear is sentative to these)
        # test with scatter plot 

#2. Multvariate normality 
        # all variables should be normally distributed 
        # see distribution  
        # chi-squared goodness of fit test    
        # can also use qqplot (step 9 below)

data = X.m_dash
norm_parameter_fit = stats.norm.fit(data)
fitted_norm = stats.norm(loc=norm_parameter_fit[0], scale=norm_parameter_fit[1])
x = np.linspace(data.min(), data.max())
plt.hist(data, normed=True, alpha=0.5)
plt.plot(x, fitted_norm.pdf(x), 'r', lw=3)
plt.show()    
# In this case a double peak distribution looks more likely 

X.hist()

#3. no Multicollinearity 
        # COND.NO = If this is above 30 
        # If pvalues for variables are high but f-value for model is fine 
        # Correlation matrix can also be checked
        # can remove correlated variables in this case 

#4. no Auto-correlation
        # when residuals are not independent from each other (e.g price dependent on previous price)
        # Durbin-Watson test can be used here 

#5. Homoscedasticity
        # constant variance assumed 
        # check scatter plots and look for increasing funnels 
```

```python
# STEP 8: CHECK OUTLIERS 
# EXPLORING FOR OUTLIERS/INFLUERCER POINTS
# Outliers plots influencers plot 

fig, ax = plt.subplots(figsize=(12,8))
fig = sma.graphics.influence_plot(modelforout, ax=ax, criterion="cooks")

data[42:49]
# OR BY INDEX :    data.loc[data.index == 584]

# indexes not label as a numpy array 
# if the 'cooked distance' is far from the rest then outlieir 

# You are looking for LARGE POINTS far away 
# as size is the computation of x and y axis 

# MUST CHECK WHAT DATA THE MODEL IS COMING FROM TO MAKE SURE INDEXES ARE THE SAME 
```

```python
# You must always Explore individual points to see if they are really outliers 
data[42:49]
```

```python
# WE CAN ALSO LOOK FOR OUTLIERS BY PLOTTING  EACH BARIABLE AGAINST THE DEPENDENT VARIABLE
# check for outliers 

fig = plt.figure(figsize=(12,8))
fig = sma.graphics.plot_partregress_grid(model, fig=fig)
```

```python
# AND EXAMINING SPECIFIC VARIABLES

fig = plt.figure(figsize=(12,8))
fig = sma.graphics.plot_regress_exog(model, "weight", fig=fig)

# outliers
# looking for blue point (real) far from red (predicted) - weight 
fig, ax = plt.subplots()
fig = sma.graphics.plot_fit(model,"weight", ax=ax)
ax.set_ylabel("Fantasy points")
ax.set_xlabel("Weight")
ax.set_title("Linear Regression")
plt.show()
```

```python
# STEP 9: CHECKING RESIDUALS 

# PLOTTING RESIDUALS 
#qqplot - to see normality of residuals 
# you want it to follow a linear line = normally distributed 

res = model.resid # residuals
fig = sma.qqplot(res)
plt.show()
```

```python
# STEP 10: COMPARE REAL/PREDICTED DATA
# This is an example of a poor prediction using linear regression 

y_test = test.avg_fantasy_points
y_pred = model.predict(add_constant(test))

predictions = model.predict(add_constant(test))
prediction_error = y_test - predictions
prediction_error

fig, ax = plt.subplots(1, 1)  
ax.scatter(y_test, y_pred)
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
plt.show()
```
